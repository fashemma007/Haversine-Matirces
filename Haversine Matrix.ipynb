```python
import pandas as pd
import numpy as np
import math
from sklearn import *
import sklearn.neighbors
```

```python
dtf = pd.read_csv("Walking Haversine.csv")
```

```python
dtf
```

```python
dtf.columns
```

```python
Places_X=dtf[['Places_X','lat_radians_X','long_radians_X']].copy()
Places_Y=dtf[['Places_Y','lat_radians_Y','long_radians_Y']].copy()
```

```python
Places_X
```

```python
Places_Y
```

```python
import haversine

dist = sklearn.neighbors.DistanceMetric.get_metric('haversine')
dist_matrix = (dist.pairwise
    (Places_X[['lat_radians_X','long_radians_X']],
     Places_Y[['lat_radians_Y','long_radians_Y']])*6371000
              )
```

```python
df_dist_matrix = (
    pd.DataFrame(dist_matrix,index=Places_X['Places_X'], 
                 columns=Places_Y['Places_Y'])
)
```

```python
df_dist_matrix
```

```python
df_dist_incols = (
    pd.melt(df_dist_matrix.reset_index(),id_vars='Places_X')
)
#Rename this column to 'distance' for relevance.
df_dist_incols= df_dist_incols.rename(columns={'value':'distance'})
```
